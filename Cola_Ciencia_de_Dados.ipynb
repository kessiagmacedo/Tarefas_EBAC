# 3. Tratamento de Dados

```python
# Remover linhas com valores nulos em coluna específica
df.dropna(subset=['coluna'], inplace=True)
```

```python
# Preencher valores nulos com a mediana
df['coluna'] = df['coluna'].fillna(df['coluna'].median())
```

```python
# Preencher valores nulos com a média
df['coluna'] = df['coluna'].fillna(df['coluna'].mean())
```

```python
# Preencher valores nulos com texto
df['coluna'] = df['coluna'].fillna('Desconhecido')
```

```python
# Substituir valores (padronização)
mapeamento = {'F': 'Female', 'f': 'Female', 'M': 'Male'}
df['coluna'] = df['coluna'].replace(mapeamento)
```

# 4. Plotar Gráficos

## 4.1 Estrutura Básica (Matplotlib)

```python
# Estrutura básica de um gráfico
plt.figure(figsize=(10, 6))  # Largura x Altura em polegadas
plt.plot(x, y)  # ou outro tipo de gráfico
plt.title('Título do Gráfico', fontsize=14)
plt.xlabel('Eixo X', fontsize=12)
plt.ylabel('Eixo Y', fontsize=12)
plt.legend()  # Exibe legenda (precisa de label no gráfico)
plt.grid(True)  # Adiciona grade
plt.tight_layout()  # Ajusta espaçamento
plt.show()
```

## 4.2 Gráfico de Dispersão (Scatter) - Correlação entre variáveis

```python
# SEABORN - Dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='coluna_x',
    y='coluna_y',
    hue='coluna_categoria',  # Cor por categoria (opcional)
    size='coluna_tamanho',   # Tamanho do ponto (opcional)
    palette='Set2',          # Paleta de cores
    alpha=0.7                # Transparência
)
plt.title('Relação entre X e Y')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# MATPLOTLIB - Dispersão
plt.figure(figsize=(10, 6))
plt.scatter(
    x=df['coluna_x'],
    y=df['coluna_y'],
    c='blue',      # Cor
    alpha=0.5,     # Transparência
    s=50           # Tamanho dos pontos
)
plt.title('Dispersão X vs Y')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

## 4.3 Gráfico de Linha (Line) - Séries temporais

```python
# SEABORN - Linha
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=df,
    x='mes',
    y='vendas',
    hue='regiao',       # Linha por categoria (opcional)
    marker='o',         # Marcador nos pontos
    palette='husl'
)
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('Mês')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=45)
plt.legend(title='Região')
plt.show()
```

```python
# MATPLOTLIB - Linha
plt.figure(figsize=(12, 6))
plt.plot(
    df['mes'],
    df['vendas'],
    color='blue',
    marker='o',
    linestyle='-',
    linewidth=2,
    label='Vendas'
)
plt.title('Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Vendas (R$)')
plt.legend()
plt.grid(True)
plt.show()
```

## 4.4 Gráfico de Barras (Bar) - Comparar categorias

```python
# SEABORN - Barras (calcula média automaticamente)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df,
    x='categoria',
    y='valor',
    hue='grupo',         # Agrupamento (opcional)
    palette='Set2',
    errorbar=None        # Remove barras de erro
)
plt.title('Média de Valor por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor Médio')
plt.xticks(rotation=45)
plt.show()
```

```python
# MATPLOTLIB - Barras
plt.figure(figsize=(10, 6))
plt.bar(
    x=df['categoria'],
    height=df['valor'],
    color='skyblue',
    edgecolor='navy'
)
plt.title('Valores por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor')
plt.xticks(rotation=45)
plt.show()
```

## 4.5 Gráfico de Contagem (Countplot) - Frequência de categorias

```python
# SEABORN - Contagem
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df,
    x='categoria',
    hue='grupo',         # Dividir por grupo (opcional)
    palette='Set2',
    order=df['categoria'].value_counts().index  # Ordenar por frequência
)
plt.title('Contagem por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```

## 4.6 Histograma - Distribuição de variável numérica

```python
# SEABORN - Histograma
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,
    x='coluna_numerica',
    bins=20,             # Número de barras
    kde=True,            # Linha de densidade
    hue='categoria',     # Dividir por categoria (opcional)
    color='skyblue'
)
plt.title('Distribuição da Variável')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

```python
# MATPLOTLIB - Histograma
plt.figure(figsize=(10, 6))
plt.hist(
    df['coluna_numerica'],
    bins=20,
    color='skyblue',
    edgecolor='black',
    alpha=0.7
)
plt.title('Distribuição')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

## 4.7 Boxplot - Outliers e estatísticas

```python
# SEABORN - Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='categoria',       # Categoria no eixo X (opcional)
    y='valor',
    hue='grupo',         # Dividir por grupo (opcional)
    palette='Set2'
)
plt.title('Distribuição de Valor por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor')
plt.show()
```

```python
# SEABORN - Boxplot simples (uma variável)
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df,
    y='coluna_numerica',
    color='salmon'
)
plt.title('Boxplot - Identificando Outliers')
plt.ylabel('Valor')
plt.show()
```

## 4.8 Mapa de Calor (Heatmap) - Correlação

```python
# SEABORN - Heatmap de Correlação
plt.figure(figsize=(12, 8))
correlacao = df.select_dtypes(include=['number']).corr()  # Só colunas numéricas
sns.heatmap(
    correlacao,
    annot=True,          # Mostrar valores
    fmt='.2f',           # Formato dos números
    cmap='coolwarm',     # Paleta de cores
    center=0,            # Centralizar cores no zero
    square=True,         # Células quadradas
    linewidths=0.5       # Linhas entre células
)
plt.title('Mapa de Correlação')
plt.tight_layout()
plt.show()
```

## 4.9 Gráfico de Pizza (Pie) - Proporções

```python
# MATPLOTLIB - Pizza
plt.figure(figsize=(8, 8))
contagem = df['categoria'].value_counts()
plt.pie(
    contagem.values,
    labels=contagem.index,
    autopct='%1.1f%%',   # Mostrar porcentagem
    colors=sns.color_palette('Set2'),
    explode=[0.05] * len(contagem),  # Separar fatias
    shadow=True
)
plt.title('Distribuição por Categoria')
plt.show()
```

## 4.10 Múltiplos Gráficos (Subplots)

```python
# Criar grade de gráficos 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico 1 - Countplot
sns.countplot(data=df, x='categoria', ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Contagem por Categoria')

# Gráfico 2 - Histograma
sns.histplot(data=df, x='valor', ax=axes[0, 1], bins=20, color='skyblue')
axes[0, 1].set_title('Distribuição de Valor')

# Gráfico 3 - Boxplot
sns.boxplot(data=df, y='valor', ax=axes[1, 0], color='salmon')
axes[1, 0].set_title('Boxplot - Outliers')

# Gráfico 4 - Dispersão
sns.scatterplot(data=df, x='x', y='y', ax=axes[1, 1], alpha=0.6)
axes[1, 1].set_title('Dispersão X vs Y')

plt.tight_layout()
plt.show()
```
